```python
import pandas as pd
import numpy as np
import glob
from natsort import os_sorted
```

```python
def generate_csv(path, output):
    files = glob.glob(path + "*")
    files = os_sorted(files)
    print(files)
    
    state_dict = {}
    for i, file in enumerate(files):
        key = "STATE " + str(i+1)
        df = pd.read_csv(file,skiprows=2, delimiter=" ")
        #print(df.columns)
        state_dict[key] = np.array(df['<Power(dBm)>']).astype('double')

    state_dict['X coord'] = np.array(df['<X(m)>']).astype('double')
    state_dict['Y coord'] = np.array(df['<Y(m)>']).astype('double')
    #print(state_dict)
    
    data = pd.DataFrame(state_dict)
    print(data)
    
    data.to_csv(output, index=False)
```

```python
generate_csv("6000 Random Rx Points (All States)/","6000_Random_Rx_Points.csv")
```

```python
generate_csv("1000 Random Test Points/", "1000_random_test_points.csv")
```

```python
generate_csv("Dataset 2 6000 Random Rx points (All states)/","Dataset_2_6000_random_points.csv")
```
